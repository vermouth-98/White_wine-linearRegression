```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
dataFrame = pd.read_csv('winequality-white.csv', sep =';')
dataFrame
```

```python
if __name__ == '__main__':
    X_train, X_test, y_train , y_test = train_test_split(dataFrame.iloc[:,:-1].values, dataFrame.iloc[:,-1].values, test_size=0.2)
    regression = LinearRegression()
   
    polygression = PolynomialFeatures()
    pipeline = Pipeline([  
        ('poly', polygression),     
        ('linear', regression)
        ])
    parameters = {
        'poly__degree': (1,2,3)
    }
    gridsearch = GridSearchCV(pipeline, parameters,n_jobs = 3, verbose = 1, scoring ='accuracy')
    gridsearch.fit(X_train, y_train)
    best_parameters = gridsearch.best_estimator_.get_params()
    print(best_parameters)
    for param_name in sorted(parameters.keys()):
        print(param_name, best_parameters[param_name])
    
```

```python
y_pred = gridsearch.predict(X_test)
for i, value in enumerate(y_pred):
    print(f'predictions: {value} real: {y_test[i]}')
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
regression.score(X_test,y_test)
scores = cross_val_score(regression,X_train,y_train, cv =10)
print(scores)
print(scores.std())
```
